```python
from matplotlib import pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
print("step 1")
```

```python
fig = plt.figure()
ax = plt.axes()
ax.grid()
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,500)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.plot(x,np.tan(x));
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x))
ax.set_title('simple plot')
ax.set_xlabel('wavelength')
ax.set_ylabel('amplitude');
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
ax.plot(x,np.sin(x),label='sin',color='y',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',color='c',linestyle='dotted')
ax.plot(x,np.sin(x+1),label='sin(x+1)',color='m',linestyle='dashdot')
ax.set_label('multiple lines')
ax.set_ylabel('period')
ax.set_ylabel('amplitude')
ax.legend()
plt.show()
```
